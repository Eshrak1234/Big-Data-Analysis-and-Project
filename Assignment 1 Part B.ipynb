Data Description & Initial Exploration

```python
import pandas as pd


df = pd.read_csv('diabetes_012_health_indicators_BRFSS2015.csv')


print("Dataset shape:", df.shape)
display(df.head())


desc = df.describe(include='all')
display(desc)


print("Missing values per column:")
print(df.isnull().sum())


print("Target class distribution (Diabetes_012):")
print(df['Diabetes_012'].value_counts())


print("\nFeature types summary:")
print(df.dtypes)
```

Clustering/Pattern Analysis

Simple K-Means Clustering (unsupervised)

```python
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import seaborn as sns
import matplotlib.pyplot as plt

features = [col for col in df.columns if col != 'Diabetes_012']
X = df[features]


scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)


kmeans = KMeans(n_clusters=3, random_state=42, n_init=10)
clusters = kmeans.fit_predict(X_scaled)

df['Cluster'] = clusters


print(df['Cluster'].value_counts())


cross_tab = pd.crosstab(df['Cluster'], df['Diabetes_012'])
print("\nCluster vs. Diabetes_012 CrossTab:")
print(cross_tab)


plt.figure(figsize=(8, 6))
sns.heatmap(cross_tab, annot=True, fmt="d", cmap="YlGnBu")
plt.title("Cluster Membership vs. Diabetes Status")
plt.xlabel("Diabetes Status (0=No, 1=Pre-diabetes, 2=Diabetes)")
plt.ylabel("Cluster")
plt.show()
```
